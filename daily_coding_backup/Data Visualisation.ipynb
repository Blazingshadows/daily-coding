```python
import matplotlib.pyplot as plt
x = [1950, 1970, 1990, 2010]
y = [2.519, 3.692, 5.263, 6.972]
plt.plot(x, y)
```

```python
time = [8, 10, 12, 14, 16, 18, 20, 22, 0, 2]

mood = [3, 4, 5, 4, 3, 6, 8, 9, 7, 6]
fig, ax = plt.subplots(2)
ax[0].set_title("time v/s mood")
ax[0].plot(time, mood)
ax[0].set(xlabel="time", ylabel="mood")
ax[1].set_title("mood v/s time")
ax[1].plot(mood, time)
ax[1].set(xlabel="mood", ylabel="time")
fig.subplots_adjust(hspace=0.5)
```

```python
index = mood.index(max(mood))
best_time = time[index]
print(best_time, max(mood))
```

```python
max_mood = max(mood)
peak_times = [t for t, m in zip(time, mood) if m == max_mood]

print(peak_times)
```

```python
plt.plot(time, mood, marker='o', linewidth=2)

plt.scatter(best_time, max_mood, s=150, color='red', zorder=5)

plt.annotate(
    'Peak mood',
    xy=(best_time, max_mood),
    xytext=(best_time, max_mood + 1),
    arrowprops=dict(arrowstyle='->')
)
plt.ylim(0, 10)

plt.grid(alpha=0.3)
plt.xlabel("Time of Day")
plt.ylabel("Mood / Energy")
plt.title("Spotify Mood Timeline")
```
