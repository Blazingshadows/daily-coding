# Series Creation Different methods

```python
import pandas as pd,numpy as np
# specify data as sequence 1
s1 = pd.Series(range(5))
type(s1)
print(s1)
```

```python
#sequence 2
s2 = pd.Series([1,2,3,4,5])
print("Series Object 2:")
print(s2)
```

```python
#sequence 3
s3 = pd.Series((1,3,5,7))
print("Series Object 3:")
print(s3)
```

```python
#specify data as ndarray
nda1 = np.arange(3,13, 3.5)
print(nda1)
ser1 = pd.Series(nda1)
ser1
```

```python
s4 = pd.Series(np.linspace(24,64,5))
print(s4)
```

```python
s5 = pd.Series(np.tile([3,5],2))
print(s5)
```

```python
# as python dictionary
stu = {'A':39,'B':41,'C':42,'D':44,'E':40,'F':36,'G':43 ,'H':33,'I':45,'J':38}
s6 = pd.Series(stu)
print(s6)
```

```python
#as scalar value
s7 = pd.Series(50000, index = ['Q1','Q2','Q3','Q4'])
print(s7)
```

```python
s9 = pd.Series(data = [32,34,36], index = ['A','B','C'])
print(s9)
```

```python
section = ['A','B','C','D','E']
contri = np.array([6700,5600,5000,5200,np.nan])
s10 = pd.Series(contri * 2, section)
print(s10)
```

```python
s11 = pd.Series([11,12.2,"HI"])
print(s11)
```

# Series Attributes
## important Series Attributes : 
* **values** : returns value array 
* **dtype** : data type of series objectn 
* **nbyte** : total no. of bytes in series object 
* **size** : no. of elements in series 
* **shape** : no. of elements including nan and missing
* **empty** : returns boolean for is series empty? 
* **hasnans** : checks if series has nan value, returns boolean
