```python
# ways to create a dictionary
student = {"Name" : "Alice", "Age" : 25, "grade" : "A"}
print(student)
```

```python
student = dict(name='Adi', grade='A', age=22)
print(student)
```

```python
data = [('name', 'Alice'), ('grade', 'A'), ('age', 22)]
student = dict(data)
print(student)
```

```python
old_dict = {'apple': 2, 'banana': 3}
new_dict = dict(old_dict)
print(new_dict)
```

```python
keys = ['name', 'grade', 'age']
values = ['Alice', 'A', 22]
student = dict(zip(keys, values))
print(student)
```

```python
# Create a dictionary where keys are numbers and values are their squares
squares = {x: x*x for x in range(5)}
print(squares) # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

```python
keys = ['key1', 'key2', 'key3']
default_value = 0
new_dict = dict.fromkeys(keys, default_value)
print(new_dict) # Output: {'key1': 0, 'key2': 0, 'key3': 0}
```
